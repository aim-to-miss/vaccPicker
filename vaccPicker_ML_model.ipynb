## Imports

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
%matplotlib inline
```

## Preprocessing

```python
X = pd.read_csv('district_data.csv')
```

```python
X
```

```python
df = pd.DataFrame(X, columns=["Total Population",	"Area",	"Population Density",	"Age (0-14)",	"Age (15-65)",	"Age (65+)",	"No of cases",	"No of hospitals"])
```

```python
df.shape
```

```python
df
```

## Model 

```python
Kmean = KMeans(n_clusters=10)
```

```python
y = Kmean.fit_predict(df)
```

```python
df['cluster'] = y + 1
```

## Display

```python
df
```

```python
fig = plt.figure(figsize =(60, 20)) 
  
# Horizontal Bar Plot 
plt.bar(X['District'], df['cluster']) 
  
# Show Plot 
plt.show() 
```
